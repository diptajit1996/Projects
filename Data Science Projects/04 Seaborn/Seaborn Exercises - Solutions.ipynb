# Exercises

** Recreate the plots below using the titanic dataframe. There are very few hints since most of the plots can be done with just one or two lines of code and a hint would basically give away the solution. Keep careful attention to the x and y labels for hints.**

** *Note! In order to not lose the plot image, make sure you don't code in the cell that is directly above the plot, there is an extra cell above that one which won't overwrite that plot!* **

```python
sns.jointplot(x='fare', y='age', data=titanic)
```

```python
sns.displot(titanic['fare'], color='red', bins=30)
```

```python
sns.histplot(titanic['fare'], color='red', bins=30)
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='rainbow')
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='rainbow', hue='class', legend=False)
# because of error comming, that it will be deprecated so we write again like these way
# If you don't put legend then also no problem, hue is important
```

```python
sns.swarmplot(x='class', y='age', data=titanic, palette='Set2')
# if you don't use hue along with palette it gives warning that it will get deprecate in seaborn future version 0.14.
```

```python
sns.swarmplot(x='class', y='age', data=titanic, palette='Set2', hue='class', legend=False)
# Here also same like boxplot
```

```python
sns.stripplot(x='class', y='age', data=titanic, palette='Set2', hue='class', legend=False)
# Here also same like boxplot
```

```python
sns.countplot(x='sex', data=titanic)
```

```python
titanic.corr(numeric_only=True)
```

```python
sns.heatmap(titanic.corr(numeric_only=True), cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(data=titanic, col='sex')
g.map(plt.hist, 'age')
```
